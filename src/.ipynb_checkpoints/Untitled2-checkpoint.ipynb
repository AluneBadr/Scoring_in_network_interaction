# Suply Chain Scoring

## Contexte et Problèmatique

La sélection des fournisseurs est un élément important de la gestion de chaîne d'approvisionnement (SCM) pour toute organisation pour réaliser leurs objectifs. Le problème a suscité un grand intérêt des praticiens. Le processus de sélection démarre avec la détermination des critères les plus importants sur une large gamme. 

Les organisations peuvent choisir des fournisseurs de n'importe où dans le monde, les pays en développement deviennent plus compétitifs compte tenu de leur faible coûts d'exploitation.  Par conséquent, au dela des critères quantitatifs, le processus de prise de décision est fortement affectés par les risques perçus. 

Le risque peut provenir de l'économie et de l'incertitude politique dans le pays du fournisseur. Les catastrophes naturelles font également partie du risque
facteur, toute catastrophe naturelle peut avoir des conséquences catastrophiques dans le réseau d'approvisionnement mondial interconnecté d'aujourd'hui. Tous ces types de perturbations peuvent nuire à la rentabilité, au cours des actions et à la réputation sur le marché de l'organisation avec des conséquences importantes à long terme. Ainsi,une bonne vue d'ensemble et une bonne compréhension de l'ensemble de la chaîne d'approvisionnement pourrait contribuer à mieux sélectionner les fournisseurs plutôt que de se concentrer uniquement sur les performances quantitatives de ces derniers.

De nombreux chercheurs appliquent des critères multiples de prise de décision (MCDM) pour la sélection des fournisseurs. Cependant, la difficulté de la sélection de fournisseur peut augmenter de manière significative, surtout dans le contexte de la mondialisation et l'externalisation qui ont principalement complexifié les chaînes d'approvisionnement. Prenons un example : 

![title](../Picture1.PNG){ https://www.atlas-mag.net/article/qu-est-ce-qu-une-chaine-d-approvisionnement }

Compte tenu des interdépendances, surtout sur la partie amont, de nombreuses sources qualifient les chaînes d'approvisionnement modernes de réseaux d'approvisionnement afin de mieux la modéliser et la comprendre. A cela s'ajoutent les multiples risques pouvant atteindre chacun de ces fournisseurs et fragiliser ainsi le réseau. C'est la raison pour laquelle le problème de sélection de fournisseur est complexe. 

Les annalyses que nous allons faire consiste à selectionner dans un ensemble de fournisseur, celui qui est le meilleur compte tenu des risques et des interdépendences relationnelles qui existent dans le réseau. Cette sélection se fait via un modèle de scoring.

# Données

Les données utilisées pour faire ces analyses sont obtenues par le biais d'une simulation. En effet, nous allons prendre le problème comme un graph d'interaction dans lequel les noeuds seront constitués de fournisseurs et de risques (perturbations). Le lien entre deux fournisseurs matérialise les échanges et le lien entre un fournisseur et une perturbation matérialise la probabilité de subir la subir.

Pour simuler les données nous allons considère un environement où il y 780 fournisseurs qui inter-agissent ensemble et 10 peerturbations potentielles :

    -TRP : risque qu'il y ait une rupture au niveau du transport
    -GEO : problème géographique (catastrophe, inondations...)
    -MP : rupture de stock des matières première
    -GRV : une grève au sein de l'entreprise fournisseur
    -CR : une crise (économique, sanitaire...)
    -FIN : un problème financier de l'entreprise fournisseur
    -LKS : lost of key suplier
    -ADM :  un problème administratif
    -INC : incendie

La réalisation de tous ces évènements peuvent avoir un impact considèrable sur les performances d'un fournisseur notamment sur la qualité du produit, sur la quantité à livrer ou sur la date de livraison.

En plus de ces perturbations, on tins aussi compte des performances des parents de chaques fournisseurs. En effet; pour chacun des fornissuers, compte tenu des interdépendances, si ces fournissuers à lui sont impactés, il sera à son tour indirectement impacté. Pour tenir compte de cela, on ajoute dans l'espérience les statistiques des parents de chaque fournisseur; Ces statistiques sont : 

    -min_score_parent : le minimum des scores de performance de l'ensemble des parents
    -max_score_parent : le maximum des scores de performance de l'ensemble des parents
    -mean_score_parent : la moyenne des scores de performance de l'ensemble des parents
    -med_score_parent : la m"diane des scores de performance de l'ensemble des parents
    -sd_score_parent : l'ecart_type des scores de performance de l'ensemble des parents.

# Création de groupes 

Nous allons essayer de partitionner les données en utiliser un algo de k-means et un algo de CAH

# K-means :

```python
silhouette = []
wcss = []
for i in range(2, 18):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(train[var])
    #append wss
    wcss.append(kmeans.inertia_)
    #append silhouette
    preds = kmeans.predict(train[var])
    silhouette.append(silhouette_score(train[var], preds))


plt.figure(figsize=(10,5))

sns.lineplot(range(2, 18), wcss,marker='o',color='red')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Pour choisir le nombe k optimal, on utilise le Elbow Method. cette méthode nous montre l'homogéneité des gropues (inertie intra-groupe) en fonction du nombre k de classes. En effet plus on augmente k , plus les gropues sont homogénes et plus on tant vers le sur-apprentissage. le k optimal est le k à partir duquel la baise de l'inertie n'est plus significative, losqu'on passe à k+1.
Au vue du graphique, il est difficile d'observer un *coude* sur la chute de la variacnce intra. Nous allons tenté une autre méthode.

# CAH : 

```python
Z = hierarchy.linkage(train[var],method='ward',metric='euclidean', optimal_ordering=True)
```

La vraie question avec le CAH c'est où couper le dendrogrmme ? Nous utilisons la méthode présentée par Nicolas Tremblay et Pierre Borgnat dans l'article *Graph Wavelets for Multiscale Community Mining*. Cette méthode s'inspire de la méthode gap statistic présentée par Hastie, Tibshirani  et Friedman.

```python
# global gap function
x_range = np.linspace(0, Z[-1, 2], num = (50))
y_range = np.zeros(x_range.size)
for j in range(x_range.size):
    y_range[j] = p.global_gap_function(x_range[j], Z)

gamma = x_range[np.argmax(y_range)]

print(gamma)
plt.plot(x_range, y_range)
plt.axvline(x = gamma, color = "purple")
plt.ylabel('Global gap function')
plt.xlabel('distance')
```

```python
plt.figure(figsize=(15, 10))
hierarchy.dendrogram(Z)
plt.ylabel('distances')
plt.axhline(y = gamma, color = "purple", linestyle = '--')
plt.axhline(y = 2, color = "green", linestyle = '--')
plt.xlabel('Suppliers')
```

Avec la méthode *maximal gap function*, le nombre optimal pour couper le dendrogram est 46; avec cette valeur nous avons 4 groupes de fournissuers. Dans la suite, nous choisirons de couper le dendrogram à un seuil un seuil un peu plus bas pour avoir 5 groubes de fournisseurs.

```python
tuned_parameters = [ {'criterion' : ['gini'],#, 'mae'],
                    'max_depth':  [8,9,10,11,12,13,14,15],
                     'min_samples_split': [2,3,4],
                    'min_samples_leaf': [2,3,4,5]}]

scores = ['accuracy']#['neg_mean_absolute_error']#,
dtree_model = DecisionTreeClassifier(random_state=0)

for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(dtree_model, tuned_parameters, cv=20)#,scoring= score)
                       
    clf.fit(X_train, Y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"% (mean, std * 2, params))
    print()
```
